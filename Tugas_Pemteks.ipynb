```python
import requests
from bs4 import BeautifulSoup

dosenSD = ['Atik Wintarti',
           'Elly Matul Imah',
           'Ibnu Febry Kurniawan',
           'Yuliani Puji Astuti',
           'Hasanuddin Al-Habib',
           'Riskyana Dewi Intan Puspitasari',
           'Fadhilah Qalbi Annisa',
           'Harmon Prayogi',
           'Ulfa Siti Nuraini',
           'Ike Fitriyaningsih',
           'Dinda Galuh Guminta',
           'Kartika Chandra Dewi',
           'Yuni Rosita Dewi']

main_url = "https://scholar.google.com/"
search_url = "scholar?hl=id&as_sdt=0%2C5&q=(Nama Dosen)&btnG="
```

```python
li = []
```

```python
for dosen in dosenSD:
  splt = dosen.split(' ')
  dosen_url = '+'.join(splt)
  url_toscrap = main_url+search_url.replace('(Nama Dosen)',dosen_url)
  response = requests.get(url_toscrap)
  soup = BeautifulSoup(response.content, 'html.parser')
  div_parent = soup.find_all('div', {'class':'gs_ri'})
  for child in div_parent:
    if child.h3.a == None:
      print(child)
      continue
    judul_artikel = child.h3.a.text
    Keterangan = child.find('div', {'class':'gs_a'})
    try:
      ket = Keterangan.text[Keterangan.text.index(', 20')+2:]
    except: ket = ''
    dicttemp = {'Nama Dosen':dosen, 'Judul Artikel':judul_artikel, 'Keterangan':ket}
    li.append(dicttemp)
    print(dicttemp)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(li)
df
```

```python
df.to_csv('ArtikelDosen.csv')
```
